# Residual Analysis

# The Q-Q Plot. A second type of diagnostic aid is the probability plot, a graph of the residuals versus the expected order statistics of the standard normal distribution. This graph is also called a Q-Q Plot because it plots quantiles of the data versus quantiles of a distribution.

# QQ Plot===============#If the data is normally distributed, the points in a Q-Q plot will lie on a straight diagonal line. Conversely, the more the points in the plot deviate significantly from a straight diagonal line, the less likely the set of data follows a normal distribution.

```python
import statsmodels.api as sm
```

```python
sm.qqplot(model.resid,line='q')
```

```python
sm.qqplot(model.resid,line='q')
plt.title('residual Plot')
```

# Residual Plot for Homoscedasticity

# Data are homoscedastic if the residuals plot is the same width for all values of the predicted DV. Heteroscedasticity is usually shown by a cluster of points that is wider as the values for the predicted DV get larger.

# Homoscedasticity, or homogeneity of variances, is an assumption of equal or similar variances in different groups being compared

```python
plt.scatter(model.fittedvalues,model.resid)
plt.show()
```

# CookDistance

# Cook's distance is the scaled change in fitted values, which is useful for identifying outliers in the X values (observations for predictor variables). Cook's distance shows the influence of each observation on the fitted response values.

# Cook's distance, Di, is used in Regression Analysis to find influential outliers in a set of predictor variables. In other words, it's a way to identify points that negatively affect your regression model.

```python
from statsmodels.graphics.regressionplots import influence_plot
import numpy as np

model_influence = model.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig =plt.subplots(figsize=(20,10))
plt.stem(np.arange(len(df)),np.round(c,3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show()
```
